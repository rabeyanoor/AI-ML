```python
def square_addition(a,b,c):
    print(f"{a} {b} {c}")
    a=a**2
    b=b**2
    return a+b+c
ans = square_addition(10,20,30)+10
print(ans)
```

```python
def square_addition(a,b,c):
    print(f"{a} {b} {c}")
    a=a**2
    b=b**2
    return a+b+c
# keyword argument
ans = square_addition(c=10,a=20,b=30)+10
print(ans)
```

```python
def square_addition(* args):
    print(args)
ans = square_addition(10,20,30,40,50)
print(ans)    
```

```python
def square_addition(* args):
    for i in  args:
        print(i)
ans = square_addition(10,20,30,40,50)
print(ans)    
```

```python
def square_addition(* args):
    sum=0
    for i in  args:
        i=i**2
        sum+=i
    return sum    
ans = square_addition(10,20,30,40,50)
print(ans)    
```

## generator

```python
lst =[x for x in  range(40)]#dataset
print(lst)
```

```python
lst =[x for x in  range(40)]#dataset
def data_loader(chunk_size,lst):
    for i in range(len(lst)):
        yield lst[i:i+chunk_size]

x=data_loader(5,lst) 
print(type(x))  
print(x)  
print(next(x))       
print(next(x))       

```

```python
lst =[x for x in  range(40)]#dataset

def data_loader(chunk_size,lst):
    for i in range(0,len(lst),chunk_size):
        yield lst[i:i+chunk_size]

x=data_loader(5,lst) 
 
print(next(x)) 
print(next(x))       


```

## Map

```python
lst =[1,2,3,4,5,6]
def square(x):
    return x**2
lst= map(square,lst)
print(lst)
```

```python
lst =[1,2,3,4,5,6]
def square(x):
    return x**2
lst= list(map(square,lst))
print(lst)
```

```python
# string

string ="hello world,welcome to the age of AI "

string1 =list(map(str.upper ,string))
print(string1)

string1= "".join(string1)
print(string1)

string1= "  ".join(string1)
print(string1)
```

```python
## map with lambda function

lst = [1,2,3,4,5,6]
lst =(map(lambda x:x**2 ,lst))

print(lst)

lst =list(map(lambda x:x**2 ,lst))

print(lst)
```

```python
lst = [1,2,3,4,5,6]
lst =list(map(lambda x:x**2  if x%2==0 else x,lst))

print(lst)
```

## Filter

```python
numbers =[x for x in range(100)]
print(numbers)

# even
even =filter(lambda x: x% 2==0,numbers)
print(even)

even =list(filter(lambda x: x% 2==0,numbers))
print(even)
```

```python
fifty_upper = list(filter(lambda x:x>50,numbers))
print(fifty_upper)
```

```python
data=[0,1,'',None,'Hello',[],[1,2],None,True,False,None]
cleaned_data =list(filter(None,data))

print(cleaned_data)
```

```python
# string

vowel_ =filter(lambda x:x in "AEIOU", string1)
print(vowel_)

vowel_ =list(filter(lambda x:x in "AEIOU", string1))
print(vowel_)
```

```python
fruits =['aaa','bb','ccc','ddddd']
filtering =list(filter(lambda x: len(x)>2 , fruits))

print(filtering)
```

## Reduce

```python
from functools import reduce

lst =[1,2,3,4,5,6]

summation=reduce(lambda x,y : x+y,lst)
print(summation)
```

```python
lst =[1,2,10,3,4,5,6]

max_value =reduce(lambda x,y :x if x>y else y ,lst)
print(max_value)
```

```python
# list convert string

lst=["hello","world","welcome","to","programming"]
print(lst)

string =reduce(lambda x,y : x+" "+y ,lst)
print(string)
```
