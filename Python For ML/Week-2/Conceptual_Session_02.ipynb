```python
class Dog():
    food="Rabeya" # class attribute
    # constructor
    def __init__(self,name,age):
      self.name=name # instance attribute
      self.age=age # instance attribute
      
dog1=Dog('ABC',20)
dog2=Dog('ABC2',40)

print(dog1.name)
print(dog1.food)
print(dog2.food)
```

```python
class Dog():
    food="Rabeya" # class attribute
    # constructor
    def __init__(self,name,age):
      self.name=name # instance attribute
      self.age=age # instance attribute

    # instance method
    def speak(self):
       print("Dog speaking",{self.name})    
      
dog1=Dog('ABC',20)
dog2=Dog('ABC2',40)

dog1.speak()
dog2.speak()

```

```python
class Dog():
    food="Rabeya" # class attribute
    # constructor
    def __init__(self,name,age):
      self.name=name # instance attribute
      self.age=age # instance attribute

    # instance method
    def speak(self):
       print("Dog speaking")   
   # instance method
    def info(self):
       print("Food",self.food)    
      
dog1=Dog('ABC',20)
dog2=Dog('ABC2',40)

dog1.info()

```

```python
class Dog():
    food="Rabeya" # class attribute
    # constructor
    def __init__(self,name,age):   # jkono name deo jabe cls , self, .....    
      self.name=name # instance attribute
      self.age=age # instance attribute

    # instance method
    def speak(self):  # instance method all attribute axcess nite parbe 
       print("Dog speaking",{self.name})   

    @ classmethod  # class method only class attribute gulo axcess korte parbe 
    def info(cls):
       print("Food", cls.food) # jkono name deo jabe cls , self, .....    
      
dog1=Dog('ABC',20)
dog2=Dog('ABC2',40)

dog1.info()

```
