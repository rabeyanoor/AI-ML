## Handling duplicates values

```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "Alice", "David", "Bob"],
    "City": ["New York", "London", "Paris", "New York", "Tokyo", "London"],
    "Score": [85, 90, 78, 70, 95, 90]
}
df=pd.DataFrame(data)


df
```

```python
df.duplicated() # full row  1 tai same
```

```python
df.duplicated().sum() # sudu 1 ta value ase duplicate
```

```python
# deleting duplicates values

df.drop_duplicates() # drop licle delete hobe
```

```python
# permanent delete

df.drop_duplicates(inplace=True) 

df
```

```python
#  name er basis e duplicate deletion

df.drop_duplicates(subset=['Name'])
```

```python
# first value k rakbe ( Alice 2 ta .first Alice k rakce)

df.drop_duplicates(subset=['City']) 
```

```python
# last value k rakbe (First Alice k delete korce. last ta k rakce)

df.drop_duplicates(subset=['City'],keep='last') 
```

```python
df.drop_duplicates(subset=['City'],keep='first')  # akhon first ta thakbe
```

# Apply Function in DF

```python
# min max scaling

mn=df['Total Marks'].min()
mx=df['Total Marks'].max()

df['Scaled Marks']=df['Total Marks'].apply(lambda x: (x-mn)/(mx-mn))
```

```python
df
```

```python
# custom built function

def grading_system(marks):
    if marks >=260:
        return 'A+'
    elif marks >=250:
        return 'A'
    else :
        return 'A-'
    

df['Grade']=df['Total Marks'].apply(grading_system)

```

```python
df
```

```python
def marking_system(df):
    a=df['Data Structure Marks']*2
    b=df['Python Marks']*3
    c=df['Algorithm Marks']*4

    return a+b+c

df['Exceptional Marks']=df.apply(marking_system,axis=1)


# apply () function er modhe kicu likle full DataFrame ta k pabe
```

```python
df
```
