```python
import numpy as np 

a=np.array([
    [1,2,3],
    [4,5,6]
])

b=np.array([
    [7,8],
    [9,10],
    [11,12]
])

d=np.dot(a,b) # dot=dot gun
print(d)
```

```python
d1=np.linalg.inv(d) # inv= inverce, linalg= linear algebra
print (d1)
```

```python
d2=np.linalg.matrix_transpose(d1)
print(d2)
```

## Array Indexing and Slicing
- 1.Create a NumPy array of shape (6, 6) with values from 1 to 36. Extract the sub-array consisting of the 3rd to 5th rows and 2nd to 4th columns.

- 2.Create a NumPy array of shape (5, 5) with random integers. Extract the elements on the border.

```python
#1
arr=np.random.randint(1,36,size=(6,6))
print(arr)
```

```python
arr2 = arr[2:5,1:4]
arr2
```

```python
arr2 = arr[2:5,1:4]
print(arr[2:5,1:4])
```

```python
# 2
arr=np.random.randint(1,36,size=(5,5))
print(arr)
```

```python
print(arr[0,:])
print(arr[-1,:])
print(arr[1:4,0])
print(arr[1:4, -1])

arr3=np.concatenate((arr[0,:],arr[-1,:],arr[1:4,0],arr[1:4,-1]))
print(arr3)
```
