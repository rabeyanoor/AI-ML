## Q1: One-Hot Encoding

## Q2: TF-IDF

## Optional Answer: CBOW vs Skip-Gram

**1. How each model predicts words**
- **CBOW (Continuous Bag of Words):** Uses the surrounding context words (a window of neighboring words) as input to predict the single target (center) word. It essentially averages/combines the context word vectors to predict the missing middle word.
- **Skip-Gram:** Works in the opposite direction — it takes the single target (center) word as input and tries to predict the surrounding context words within a given window.

**2. Which model performs better on small datasets**
- **CBOW** generally performs better and trains faster on **small datasets**, because it smooths over a lot of distributional information by averaging context words, which helps when there isn't much training data.

**3. Which model is better for learning rare words**
- **Skip-Gram** tends to work better with **rare words / infrequent words**, since it generates multiple training pairs (target → each context word) for every occurrence of a word, giving rare words more chances to have their representation updated.

**4. One practical advantage of Word2Vec over One-Hot Encoding**
- Word2Vec produces **dense, low-dimensional vectors that capture semantic relationships** between words (e.g., similar words have similar vectors, and relationships like *king − man + woman ≈ queen* can be captured). One-Hot Encoding, in contrast, produces **sparse, high-dimensional vectors with no notion of similarity** between words — every word is equally "different" from every other word.
