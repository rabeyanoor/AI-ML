```python
def remove_punc(text):
  for char in exclude:
    text = text.replace(char,"")

  return text
```

# Stopwords removal

```python
text = """
Yesterday, I was working on a Machine Learning project in PHITRON, and I was trying to understand how the model is performing on the dataset. The dataset is quite large, and it contains a lot of reviews that are written by users from different countries.

I noticed that the model is not performing well for some of the samples, especially those that are very short or have too many stopwords in them. Because of this, I decided to clean the text data and remove unnecessary words that do not contribute much to the meaning of the sentence.

While I was doing this, I also checked how the preprocessing steps like tokenization, lowercasing, and punctuation removal are affecting the dataset. It was interesting to see that after removing stopwords, the text becomes more meaningful and easier for the model to understand.

In addition to that, I was also reading some documentation from https://scikit-learn.org/stable/ and watching tutorials on https://www.youtube.com while comparing different NLP techniques. The goal is to improve the accuracy of the model and make it more robust for real-world applications.

Overall, I think that removing stopwords is an important step in the NLP pipeline because it helps reduce noise and improve performance of machine learning models.
"""
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
def remove_stopwords(text):
  words = text.split()
  cleaned = []

  for word in words:
    # print(word)
    word = word.lower().strip(string.punctuation)

    if word and word not in stop_words:
      cleaned.append(word)

  return " ".join(cleaned)


    # print(word)


```

```python
remove_stopwords(text)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

# Emoji removal

```python
text_with_emoji = '''Yesterday I completed my Machine Learning project on sentiment analysis ðð¥.
The model performed really well on the IMDB dataset ðð¯.
I learned a lot about preprocessing steps like tokenization, stopwords removal, and stemming ðð¡.
However, debugging errors was quite frustrating ðð».
Overall, the experience was amazing and I feel more confident now ðð.'''
```

```python
import re

def remove_emoji(text):
    emoji_pattern = re.compile(
        "["
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F1E0-\U0001F1FF"  # flags
        "\U00002700-\U000027BF"  # dingbats
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)
```

```python
clean_text = remove_emoji(text_with_emoji)
print(clean_text)
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = "walk walks walking walked"
stem_words(sample)
```

# Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
sentence = "He was running and eating at same time. He has bad habit of swimming after playing long hours in the Sun."
```

```python
punctuations="?:!.,;"
```

```python
sentence_words = nltk.word_tokenize(sentence)
```

```python
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)
```

```python
sentence_words
```

```python
print("{0:20}{1:20}".format("Word","Lemma"))
```

```python
for word in sentence_words:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
