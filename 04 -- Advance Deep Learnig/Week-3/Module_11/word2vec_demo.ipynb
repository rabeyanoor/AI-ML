```python
import gensim.downloader as api
import torch
import torch.nn as nn

# 1. Automatically download, extract, and load the 300d Google News model
print("Loading Google News Word2Vec...")

word2vec_model = api.load('word2vec-google-news-300')

```

```python
# 1. Download the pre-trained weights from a working Hugging Face mirror

!wget -c -O GoogleNews-vectors-negative300.bin.gz "https://huggingface.co/rajnishrajput12/finber/resolve/main/GoogleNews-vectors-negative300.bin.gz"

import gensim
from gensim.models import KeyedVectors
import warnings
warnings.filterwarnings('ignore') # Hides the Gensim vstack FutureWarning from your output
```

```python
# 2. Load the embeddings (limiting to 500,000 words to save RAM)

print("Loading Word2Vec model into memory...")

model = KeyedVectors.load_word2vec_format('GoogleNews-vectors-negative300.bin.gz', binary=True, limit=500000)
print("Model loaded successfully!\n")
print("-" * 50)
```

```python
# 3. Extracting the vector array for a specific word
print("Vector array for 'cricket' (first 10 dimensions):")

print(model['king'][:10])
```

```python
# 4. Semantic Similarity (Most Similar)

print("\nMost similar to 'man':")
for word, score in model.most_similar('man', topn=5):
    print(f" - {word}: {score:.4f}")

```

```python
# . Semantic Similarity (Most Similar)

print("\nMost similar to 'man':")
for word, score in model.most_similar('Bangladesh', topn=5):
    print(f" - {word}: {score:.4f}")
```

```python
print("\nMost similar to 'facebook':")
for word, score in model.most_similar('facebook', topn=5):
    print(f" - {word}: {score:.4f}")
```

```python
model.most_similar("Football")
```

```python
# 5. Direct Cosine Similarity Scoring

print("\nSimilarity between 'man' and 'woman':", model.similarity('man', 'woman'))
print("Similarity between 'man' and 'PHP':", model.similarity('man', 'PHP'))
```

```python
# 6. Outlier Detection
print("\nDoesn't match ['PHP', 'java', 'monkey']:")
print(" ->", model.doesnt_match(['PHP', 'java', 'monkey']))
```

```python
# 7. Deep Semantic Vector Math (Analogies)
# king - man + woman = ?

print("\nAnalogy: king - man + woman = ")

vec_king = model['king'] - model['man'] + model['woman']

for word, score in model.most_similar([vec_king], topn=3):
    print(f" - {word}: {score:.4f}")
```
