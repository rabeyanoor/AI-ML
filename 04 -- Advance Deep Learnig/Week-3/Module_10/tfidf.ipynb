```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "students watch youtube",
    "youtube watch youtube",
    "students write comments",
    "youtube write comments"
]
```

```python
# ==========================================
# 1. Create Dataset
# ==========================================
df = pd.DataFrame({ "text" : corpus
    ,
    "output": [1, 1, 0, 0]
})

print("Original DataFrame:")
print(df)


```

```python
#  ==========================================
# 2. Create TF-IDF Vectorizer
# ==========================================
tfidf = TfidfVectorizer()
```

```python
# ==========================================
# 3. Learn Vocabulary + Compute TF-IDF
# ==========================================
X = tfidf.fit_transform(df['text'])

```

```python
# ==========================================
# 4. Convert Sparse Matrix to DataFrame
# ==========================================
tfidf_df = pd.DataFrame(
    X.toarray(),
    columns=tfidf.get_feature_names_out()
)

print("\nTF-IDF Matrix:")
print(tfidf_df)
```

```python
# ==========================================
# 5. Print Vocabulary
# ==========================================
print("\nVocabulary:")
print(tfidf.vocabulary_)
```

```python
# ==========================================
# 6. Print Feature Names
# ==========================================
print("\nFeature Names:")
print(tfidf.get_feature_names_out())

```

```python
# ==========================================
# 7. Print IDF Values
# ==========================================
print("\nIDF Values:")
for word, idf in zip(tfidf.get_feature_names_out(), tfidf.idf_):
    print(f"{word:10} -> {idf:.4f}")

```

```python
# ==========================================
# 8. Transform New Sentence (Inference)
# ==========================================
new_text = ["people watch comment"]

new_vector = tfidf.transform(new_text)

print("\nNew Sentence:")
print(new_text[0])

print("\nTF-IDF Vector for New Sentence:")
print(pd.DataFrame(
    new_vector.toarray(),
    columns=tfidf.get_feature_names_out()
))

```

```python
# ==========================================
# 9. Example with OOV Word
# ==========================================
new_text_oov = ["people watch youtube"]

new_vector_oov = tfidf.transform(new_text_oov)

print("\nSentence with New Word:")
print(new_text_oov[0])

print("\nTF-IDF Vector:")

print(pd.DataFrame(
    new_vector_oov.toarray(),
    columns=tfidf.get_feature_names_out()
))
```
