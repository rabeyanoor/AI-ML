```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
```

```python
# ==========================================
# 1. Documents (Corpus)
# ==========================================
corpus = [
    "students watch youtube youtube youtube",
    "youtube watch youtube",
    "students write comments",
    "youtube write comments"
]
```

```python
# create bag of words model

cv = CountVectorizer(binary=True)
```

```python
# learn vocabulary and transfrom documnent

X = cv.fit_transform(corpus)
```

```python
#show vocabulary

cv.get_feature_names_out()
```

```python
# show bag of words matrix

X.toarray()
```

```python
bow_df = pd.DataFrame(
    X.toarray(),
    columns = cv.get_feature_names_out(),
    index = ["D1","D2","D3","D4"]
)
bow_df
```
