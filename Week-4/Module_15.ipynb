### 2D line plot


- ` How a value changes over time.`
- ` Time series data.`
- ` Example--> Business data, temperatur data.`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# for one student

hours =[2,3,4,1,5,7,3] # x
days=[1,2,3,4,5,6,7] # y

plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week.')


plt.plot(days,hours)
```

```python
# for multipol student

hours_1 =[2,3,4,1,5,7,3] 
hours_2=[1,4,3,5,7,2,1]
days=[1,2,3,4,5,6,7] 


plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week.')


plt.plot(days,hours_1)
plt.plot(days,hours_2)
```

```python
df=pd.read_csv('enrollment_data.csv')  # x er jaigay df or jkono kicu dea jabe
df
```

```python
plt.xlabel('Years')
plt.ylabel('count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
# adding labels using plt.legend()

plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend()
```

```python
# plt.legent() location
# upper right

plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend(loc='upper right')
```

```python
# lower right

plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend(loc='lower right')
```

```python
# best 


plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend(loc='best')
```
