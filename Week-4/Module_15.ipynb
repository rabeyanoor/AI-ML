### 2D line plot


- ` How a value changes over time.`
- ` Time series data.`
- ` Example--> Business data, temperatur data.`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# for one student

hours =[2,3,4,1,5,7,3] # x
days=[1,2,3,4,5,6,7] # y

plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week.')


plt.plot(days,hours)
```

```python
# for multipol student

hours_1 =[2,3,4,1,5,7,3] 
hours_2=[1,4,3,5,7,2,1]
days=[1,2,3,4,5,6,7] 


plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week.')


plt.plot(days,hours_1)
plt.plot(days,hours_2)
```

```python
df=pd.read_csv('enrollment_data.csv')  # x er jaigay df or jkono kicu dea jabe
df
```

```python
plt.xlabel('Years')
plt.ylabel('count of Enrollment')
plt.title('Student Enrollment over 8 years')

plt.plot(df['Year'],df['Programming'])
plt.plot(df['Year'],df['Digital Marketing'])
```

```python
# adding labels using plt.legend()

plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend()
```

```python
# plt.legent() location
# upper right

plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend(loc='upper right')
```

```python
# lower right

plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend(loc='lower right')
```

```python
# best 


plt.plot(df['Year'],df['Programming'],label='Programming')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing')

plt.legend(loc='best')
```

```python
# adding some visualization

plt.xlabel('Years')
plt.ylabel('count of Enrollment')
plt.title('Student Enrollment over 8 years')


plt.plot(df['Year'],df['Programming'],label='Programming',color='#fa0234',linestyle='dashed') # google a hex color lekha search koro 
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing',color='black',linewidth=3) # search plt linestyle

plt.legend()
```

```python
# adding marker

plt.plot(df['Year'],df['Programming'],label='Programming', marker='x')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing',marker='o',markersize='5')

plt.legend()
```

```python
# grid

plt.plot(df['Year'],df['Programming'],label='Programming', marker='x')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing',marker='o',markersize='5')

plt.legend()
plt.grid()
```

```python
# grid

plt.plot(df['Year'],df['Programming'],label='Programming', marker='x')
plt.plot(df['Year'],df['Digital Marketing'],label='Digital Marketing',marker='o',markersize='5')

plt.legend()
plt.grid()
plt.show() # eta likle onno jkono platform a eta show korbe. ata use na korle graph show na korteo pare
```

### Usage of Scatter Plots

```python
# Scatter plots


students = pd.read_csv('student_IQdata.csv')
students
```

```python
# Relation between Study Hours and IQ Score --> positive  correlation


plt.xlabel('Study Hours')
plt.ylabel('IQ Score')
plt.title('Relation between Study Hours and IQ Score')


plt.scatter(students['Study_Hour'],students['IQ_Score'])
```

```python
# Relation between Shoe size and IQ Score --> no correlation


plt.xlabel('Shoe_Size')
plt.ylabel('IQ Score')
plt.title('Relation between Shoe_Size and IQ Score')


plt.scatter(students['Shoe_Size'],students['IQ_Score'])
```

```python
# Relation between Chilling_Hours and IQ Score --> 


plt.xlabel('Chilling_Hours')
plt.ylabel('IQ Score')
plt.title('Relation between Chilling_Hours and IQ Score')


plt.scatter(students['Chilling_Hours'],students['IQ_Score'],marker='x',color='red')
```

### Bar Chart

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# gender wise

gender_group = df.groupby('Gender').size()

gender_group
```

```python
gender_group = df.groupby('Gender').size()

gender_group.index
```

```python
gender_group.values
```

```python
plt.bar(gender_group.index, gender_group.values)
```

```python
plt.bar(gender_group.index,gender_group.values,color=['orange','red'],edgecolor='black')


plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gender among students')
```

```python
# Bar chart on real data

df=pd.read_csv('student_data.csv')
df
```

```python
status  =df.groupby('CompletionStatus').size()

plt.bar(status.index, status.values, color=['green','orange','red'], edgecolor='black')

plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.title('Bar Chart of Course completion')
plt.show()
```

```python
# location

status=df.groupby('Location').size()

plt.bar(status.index,status.values,edgecolor='red')

plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar Chart of location')
plt.show()
```
