### 2D line plot


- ` How a value changes over time.`
- ` Time series data.`
- ` Example--> Business data, temperatur data.`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
# for one student

hours =[2,3,4,1,5,7,3] # x
days=[1,2,3,4,5,6,7] # y

plt.plot(days,hours)
```

```python
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week.')


plt.plot(days,hours)
```

```python
# for multipol student

hours_1 =[2,3,4,1,5,7,3] 
hours_2=[1,4,3,5,7,2,1]
days=[1,2,3,4,5,6,7] 


plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Trend of a student study time over a week.')


plt.plot(days,hours_1)
plt.plot(days,hours_2)
```
