```python
import numpy as np 

a=np.array([
    [1,2,3],
    [4,5,6]
])

b=np.array([
    [7,8],
    [9,10],
    [11,12]
])

d=np.dot(a,b) # dot=dot gun
print(d)
```

```python
d2=np.linalg.matrix_transpose(d1)
print(d2)
```
