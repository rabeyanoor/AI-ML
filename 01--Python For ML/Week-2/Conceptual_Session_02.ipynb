### Inheritance

```python
## Inherit (single inheritance)
## Parent Class / Base class / Super class
class Car:
  def __init__(self, windows, door, enginetype) -> None:
     self.windows = windows
     self.door = door
     self.enginetype = enginetype


  def drive(self):
    print(f"The person will drive the {self.enginetype} card")
```

```python
# Derived/Child/ Subclass

class Tesla(Car):
    def __init__(self,windows,door,enginetype, is_selfDriving):
        super().__init__(windows,door,enginetype)
        self.is_selfDriving= is_selfDriving

teslal= Tesla(4,4,"AB",True)
print(teslal.windows)
```

```python
# Multiple Inheritance
# When a class inherits from more than one base class


## Base class 1
class Animal:
    def __init__(self,name) -> None:
        self.name=name

    def eat(self):
        print(f"{self.name} is eating from Animal Base Class")


## Base class 2
class Pet:
    def __init__(self,owner) -> None:
        self.owner= owner


# Derive class
class Dog(Animal,Pet):
    def __init__(self,name,owner)  -> None:
        Animal. __init__(self.name)
        Pet.__init__(self.owner)

dog1=Dog("ABC","DEF")
dir(dog1)



        
```

```python
## Multiple Inheritance
## When a class inherits from more than one base class

## Base class 1
class Animal:
  def __init__(self, name) -> None:
    self.name = name

  def eat(self):
    print(f"{self.name} is eating from Animal Base Class")

## Base class 2
class Pet:
  def __init__(self, owner) -> None:
     self.owner = owner


# Derive class
class Dog(Animal, Pet):
  def __init__(self, name, owner) -> None:
    Animal.__init__(self, name)
    Pet.__init__(self, owner)


dog1 = Dog("ABC", "DEF")
dir(dog1)
```
