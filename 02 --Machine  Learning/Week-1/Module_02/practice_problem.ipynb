```python
weights = [2.5,3.0,3.2,3.3,3.4,3.5,3.6,3.9,4.0,4.5]
df = pd.DataFrame({"w": weights})

df
```

```python
import numpy as np

p25 = np.percentile(df["w"], 25) # [same]
p50 = df["w"].quantile(0.50) # [same]  
p75 = df["w"].quantile(0.75)
iqr = p75 - p25

p25, p50, p75, iqr
```

```python
# Outlier limit
lower = p25 - 1.5*iqr
upper = p75 + 1.5*iqr
print("Q3 Lower Fence:", lower)
print("Q3 Upper Fence:", upper)
```

```python
cats = pd.Series([2.5, 3.0, 3.2, 3.3, 3.4, 3.5, 3.6, 3.9, 4.0, 4.5])

Q1 = cats.quantile(0.25)
Q2 = cats.quantile(0.50)
Q3 = cats.quantile(0.75)
IQR = (Q3 - Q1).round(2)

Q1, Q2, Q3, IQR
```

```python
## Outlier Range
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

lower, upper
```

```python
# Z-Score (x = 65)
mu = 50
sigma = 10
x = 65

z = (x - mu) / sigma
z
```

```python
# x for Z = -2.5
z = -2.5
x = mu + z * sigma
x
```
