```python
import pandas as pd

marks = [50, 60, 65, 70, 75, 80, 85, 90]
df = pd.DataFrame({"marks": marks})
df
```

```python
mean_ = df["marks"].mean()
median_ = df["marks"].median()
mode_ = df["marks"].mode()[0]

mean_, median_, mode_
```

```python
weights = [2.5,3.0,3.2,3.3,3.4,3.5,3.6,3.9,4.0,4.5]
df = pd.DataFrame({"w": weights})

df
```

```python
import numpy as np

p25 = np.percentile(df["w"], 25) # [same]
p50 = df["w"].quantile(0.50) # [same]  
p75 = df["w"].quantile(0.75)
iqr = p75 - p25

p25, p50, p75, iqr
```

```python
# Outlier limit
lower = p25 - 1.5*iqr
upper = p75 + 1.5*iqr
print("Q3 Lower Fence:", lower)
print("Q3 Upper Fence:", upper)
```

```python
mu = 50
sigma = 10

# a) Z score for x=65
z_65 = (65 - mu)/sigma
print("\nQ4 Z for x=65:", z_65)

# b) x from Z = -2.5
x_neg = mu + (-2.5)*sigma
print("Q4 x for Z=-2.5:", x_neg)
```
